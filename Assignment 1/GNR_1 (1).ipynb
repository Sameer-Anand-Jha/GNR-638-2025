```python
def plot_tsne(pkl_file, labels, categories):
    import pickle
    import matplotlib.pyplot as plt
    from sklearn.manifold import TSNE

    # Load features from .pkl file
    with open(pkl_file, 'rb') as f:
        features = pickle.load(f)

    # Repeat labels to match the number of features
    num_samples_per_category = len(features) // len(labels)
    expanded_labels = []
    for label in labels:
        expanded_labels.extend([label] * num_samples_per_category)

    # Handle any remaining samples (edge case)
    while len(expanded_labels) < len(features):
        expanded_labels.append(labels[-1])  # Assign the last label

    # Check label length
    if len(expanded_labels) != len(features):
        print(f"Error: Expanded labels ({len(expanded_labels)}) and features ({len(features)}) still do not match.")
        return

    # Run t-SNE
    tsne = TSNE(n_components=2, random_state=42, perplexity=30)
    reduced_features = tsne.fit_transform(features)

    # Plot t-SNE
    plt.figure(figsize=(10, 7))
    scatter = plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=expanded_labels, cmap='tab20', alpha=0.7)
    cbar = plt.colorbar(scatter, ticks=range(len(categories)))
    cbar.ax.set_yticklabels(categories)
    plt.title("t-SNE Visualization")
    plt.xlabel("t-SNE Component 1")
    plt.ylabel("t-SNE Component 2")
    plt.savefig("tsne_visualization" + "300.png")
    plt.show()
CATEGORIES = ['agricultural', 'airplane', 'baseballdiamond', 'beach', 'buildings',
              'chaparral', 'denseresidential', 'forest', 'freeway', 'golfcourse',
              'harbor', 'intersection', 'mediumresidential', 'mobilehomepark', 'overpass',
              'parkinglot', 'river', 'runway', 'sparseresidential', 'storagetanks', 'tenniscourt']

CATE2ID = {category: idx for idx, category in enumerate(CATEGORIES)}  # Map category to numeric labels
numeric_labels = list(CATE2ID.values())  # Numeric labels [0, 1, ..., 20]

plot_tsne("train_image_feats_300.pkl", numeric_labels, CATEGORIES)
```
